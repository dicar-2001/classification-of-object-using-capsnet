```python
import trimesh

mesh = trimesh.load(r'ModelNet40\airplane\train\airplane_0002.off')
```

```python
points, _ = trimesh.sample.sample_surface(mesh, count=10000)
```

```python
import numpy as np
from skimage import measure

resolution = 32 # a reflechir

grid_size = (resolution, resolution, resolution)
grid = np.zeros(grid_size, dtype=bool)
points_normalized = (points - points.min(axis=0)) / (points.max(axis=0) - points.min(axis=0))
points_voxels = (points_normalized * (resolution - 1)).astype(int)

grid[points_voxels[:, 0], points_voxels[:, 1], points_voxels[:, 2]] = True
```

```python
import numpy as np
import plotly.graph_objs as go

x, y, z = np.where(grid)
scatter = go.Scatter3d(
    x=x, y=y, z=z,
    mode='markers',
    marker=dict(
        size=5,
        color='blue',
        symbol='square'
    )
)
layout = go.Layout(
    title='3D Voxel Grid Visualization',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

fig = go.Figure(data=[scatter], layout=layout)
fig.show()
```
